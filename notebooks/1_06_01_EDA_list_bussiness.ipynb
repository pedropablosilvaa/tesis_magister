```python
def count_clases(df, column_name):
    sns.set_theme(style="darkgrid")
    ax = sns.countplot(y=column_name, data=df)
    sns.set(rc = {'figure.figsize':(25,8)})
    #for item in ax.get_xticklabels():
    #    item.set_rotation(90)

count_clases(business_location,'Comuna_l')
```

```python
count_clases(business_location,'Ciudad')
```

```python
count_clases(business_location,'Actividad económica')
```

```python
business_location_clean = business_location[business_location['lon'].notna()]
business_location_clean = business_location[business_location['lat'].notna()]
business_location_gdp = gpd.GeoDataFrame(
    business_location_clean, geometry=gpd.points_from_xy(business_location_clean.lon, business_location_clean.lat))
```

```python
fig, ax = plt.subplots(figsize=(12,12))
business_location_gdp.plot(color='cyan',ax=ax,  alpha=0.4)
ax.ticklabel_format(style='plain')
ax.set_title('layer business_location')
cx.add_basemap(ax, source=cx.providers.CartoDB.Voyager)
plt.tight_layout()
plt.show()
```

```python
export_geojson = '../data/GIS/intermedios/business_location.geojson'
business_location_gdp.to_file(export_geojson, driver='GeoJSON')
```
