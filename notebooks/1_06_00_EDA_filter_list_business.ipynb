```python
sii = df_with_location.applymap(str)
pd.unique(sii['Calle'])
```

```python
import time
start_time = time.time()
x = []
y = []
I = 0
for i in range(len(sii)):
    if sii['Ciudad'].iloc[i] == 'nan' and sii['Numero'].iloc[i] == 'nan':
        direccion = sii['Calle'].iloc[i] + ', ' + sii['Comuna_r'].iloc[i] + ', ' + 'Los Lagos Region'
    elif sii['Ciudad'].iloc[i] == 'nan':
        direccion = sii['Calle'].iloc[i] + ', ' + sii['Comuna_r'].iloc[i] + ', ' + 'Los Lagos Region'
    else:
        direccion = sii['Calle'].iloc[i] + ' ' + sii['Numero'].iloc[i] + ', ' + sii['Ciudad'].iloc[i] + ', ' + 'Los Lagos Region'
    
    if len(geocoder.osm(direccion)) != 0:
        x.append(geocoder.osm(direccion).osm['x'])
        y.append(geocoder.osm(direccion).osm['y'])
    else:
        x.append(0)
        y.append(0)
time_execution = time.time() - start_time
print(f"Se codificaron {len(sii)} registros en un tiempo de ejecucion de {time_execution} s.")
```

```python
sii['lon'] = x
sii['lat'] = y
```
